```python
def rgb2gray(rgb):
    if rgb.max()>1:
        rgb=rgb[...,:3]/255.     
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
def calcHist(raw): 
    gray_cur=rgb2gray(raw)
    hist=np.zeros(255)
    unique, counts = np.unique(gray_cur*255, return_counts=True)
    for i,rank in enumerate(unique):
        hist[int(rank)]=counts[i]
    return hist
```

```python
def draw4(curImage):
    hist=calcHist(curImage)
    pdf=hist/np.sum(hist)
    cdf=[np.sum(pdf[:i]) for i in range(256)]
    plt.figure(1,figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    
    plt.subplot(221)
    plt.imshow(curImage)

    plt.subplot(222)
    plt.grid(True)
    plt.title('Histogram of Image')
    plt.xlabel('gray scale rank level')
    plt.ylabel('number of pixels')
    plt.plot(hist)

    plt.subplot(223)
    plt.grid(True)
    plt.title('PDF of Image')
    plt.xlabel('gray scale rank level')
    plt.ylabel('Probability density function')
    plt.plot(pdf)

    plt.subplot(224)
    plt.grid(True)
    plt.title('CDF of Image')
    plt.xlabel('gray scale rank level')
    plt.ylabel('Cumulative distribution function')
    plt.plot(cdf)

    plt.tight_layout()
```

```python
draw4(image_crowd)
```
